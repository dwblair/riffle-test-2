```python
%matplotlib inline 
# above line allows for showing plots in a cell easily
import matplotlib # the main plotting library
import numpy as np # allows for arrays
import matplotlib.pyplot as plt # getting specific plotting functions
import urllib # URL import functions
import pandas as pd # nice data utilities
```

```python
q=pd.read_table('./data/overnight.csv', sep=',', header=0, names=['a','b','c','d','e'])
```

```python
a=(q.a - q.a[0])/3600; c=q.c  #use unix seconds column; but start from t=0
```

```python
plt.plot(a,c);plt.xlabel('time (hours)'); plt.ylabel('temp (C)'); plt.savefig('plot.svg'); plt.savefig('plot.png') #plot, and add x and y labels
```

```python
plt.plot(a,c);plt.xlabel('time (hours)'); plt.ylabel('temp (C)'); 
plt.xlim(3,7), plt.ylim(-8,2), plt.savefig('plot_zoom1.svg'); 
plt.savefig('plot_zoom2.png') #plot, and add x and y labels
```
